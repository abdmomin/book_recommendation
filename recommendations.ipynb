```python
csv_book_mapping = {}
with open('book_id_map.csv', 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        csv_id, book_id = line.strip().split(',')
        csv_book_mapping[csv_id] = book_id
```

```python
overlap_users = set()

with open('goodreads_interactions.csv', 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        user_id, csv_id, _, ratings, _ = line.split(',')
        
        if user_id in overlap_users:
            continue
        
        try:
            ratings = int(ratings)
        except ValueError:
            continue
            
        book_id = csv_book_mapping[csv_id]
        
        if book_id in liked_books and ratings >= 4:
            overlap_users.add(user_id)
```

```python
rec_line = []
with open('goodreads_interactions.csv', 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        user_id, csv_id, _, ratings, _ = line.split(',')
        
        if user_id in overlap_users:
            book_id = csv_book_mapping[csv_id]
            rec_line.append([user_id, book_id, ratings])
```

```python
len(overlap_users)
```

```python
len(rec_line)
```

```python
import pandas as pd

recs = pd.DataFrame(rec_line, columns=['user_id', 'book_id', 'ratings'])
recs['book_id'] = recs['book_id'].astype(str)
```

```python
top10_recs = recs['book_id'].value_counts().head(10)
top10_recs = top10_recs.index.values
```

```python
top10_recs
```

```python
book_titles = pd.read_json('book_titles.json')
```

```python
book_titles['book_id'] = book_titles['book_id'].astype(str)
```

```python
book_titles.head()
```

```python
book_titles[book_titles['book_id'].isin(top10_recs)]
```

```python
all_recs = recs['book_id'].value_counts()
```

```python
all_recs = all_recs.to_frame().reset_index()
```

```python
all_recs.columns = ['book_id', 'book_count']
all_recs
```

```python
all_recs = all_recs.merge(book_titles, how='inner', on='book_id')
```

```python
all_recs.head()
```

```python
all_recs['score'] = all_recs['book_count'] * (all_recs['book_count'] / all_recs['ratings'])
```

```python
all_recs.sort_values('score', ascending=False).head(10)
```

```python
popular_recs = all_recs[all_recs['book_count'] > 75].sort_values('score', ascending=False)
```

```python
def make_clickable(link):
    return '<a target="_blank", href="{}">Goodreads</a>'.format(link)

def show_image(link):
    return '<img src="{}", width=50></img>'.format(link)

popular_recs[~popular_recs['book_id'].isin(liked_books)].head(10).style.format({'url': make_clickable, 'cover_image': show_image})
```
